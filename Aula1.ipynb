# Desafio 07: Deixar os grÃ¡ficos bonitÃµes. (Matplotlib.pyplot)

```python
explode = (0.1, 0)
plt.pie(dados['tratamento'].value_counts(), labels= dados['tratamento'].unique(), explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Tratamento')
plt.axis('equal')
plt.show()
```

```python
plt.bar(dados['tratamento'].unique(), dados['tratamento'].value_counts(), width=0.8, bottom=0, align='center', data=None)
plt.title('Tratamento')
plt.ylabel('Qty')
plt.xlabel('Class')
plt.show()
```
